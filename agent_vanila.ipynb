## 4. Human-in-the-Loop

  - ë¡œë´‡ì—ê²Œ ì‹¤ì œ ëª…ë ¹ì„ ë³´ë‚´ì•¼ í•˜ëŠ” ê²½ìš° ì•ˆì „ì„ ìœ„í•´ ì‚¬ìš©ìì˜ ìŠ¹ì¸ì„ ë°›ì•„ì•¼ í•¨.

```python
from __future__ import annotations

from dataclasses import dataclass
import re
from typing import Optional


@dataclass(frozen=True)
class HitlDecision:
    requires_approval: bool
    reason: str
    proposed_task: Optional[str] = None


_ACTION_PATTERNS = [
    # ëª…ì‹œì ìœ¼ë¡œ "ëª…ë ¹"ì„ ë‚´ë¦¬ë¼ê³  í•˜ëŠ” ê²½ìš°
    r"(ë¡œë´‡|ê¸°ê¸°|ì¥ì¹˜).{0,20}(ì—ê²Œ|ì—)\s*.+?(ëª…ë ¹|ì§€ì‹œ)\s*(í•˜|ë‚´ë¦¬)",
    # ì‚¬ìš©ìê°€ ë¡œë´‡ì—ê²Œ ë§í•˜ë¼ëŠ” ì¸ìš©/ë°œí™” ì§€ì‹œ
    r"(ë¡œë´‡|ì•„ë‹´|ì´ë¸Œ|ì• í”Œ).{0,20}(ì—ê²Œ|ì—)\s*[\"â€œ].+?[\"â€]\s*(ë¼ê³ |ì´ë¼)\s*(ë§í•˜|ëª…ë ¹í•˜|ì§€ì‹œí•˜)",
    # í•œêµ­ì–´ ëª…ë ¹í˜•/ìš”ì²­í˜•(ë¬¸ì„œ ë§¤ë‰´ì–¼ í†¤)
    r"(í•˜ì‹­ì‹œì˜¤|í•˜ì‹œì˜¤|í•˜ì„¸ìš”)\b",
    r"(ì‹¤í–‰í•´|ì¼œì¤˜|êº¼ì¤˜|ëˆŒëŸ¬|ì—´ì–´|ë‹«ì•„|ì¬ë¶€íŒ…|ë¦¬ë¶€íŒ…|ì´ˆê¸°í™”)\b",
    # troubleshooting/ì§„ë‹¨ì„ ì‹¤ì œ ìˆ˜í–‰í•˜ë¼ëŠ” ê²½ìš°
    r"(ì§„ë‹¨|í…ŒìŠ¤íŠ¸|ìº˜ë¦¬ë¸Œë ˆì´ì…˜|í•‘\s*í…ŒìŠ¤íŠ¸|POST).{0,10}(ì‹¤í–‰|ìˆ˜í–‰)\b",
]

_ACTION_RE = re.compile(
    "|".join(f"(?:{p})" for p in _ACTION_PATTERNS), re.IGNORECASE | re.DOTALL)


def decide_hitl(response_text: str) -> HitlDecision:
    """
    Regexë¥¼ í†µí•´ ì—ì´ì „íŠ¸ ì‘ë‹µì´ 'ë¡œë´‡/ì¥ì¹˜ì— ëŒ€í•œ ì‹¤ì œ í–‰ë™'ì„ ìœ ë°œí•˜ëŠ”ì§€ ê°„ë‹¨íˆ íŒë³„.
    - requires_approval=True ì´ë©´ Human-in-the-Loop ìŠ¹ì¸ ì´í›„ì—ë§Œ ë‹¤ìŒ ë‹¨ê³„ ì‹¤í–‰
    """
    text = (response_text or "").strip()
    if not text:
        return HitlDecision(False, reason="empty_response")

    if _ACTION_RE.search(text):
        return HitlDecision(
            True,
            reason="actionable_robot_command_detected",
            proposed_task=text,
        )

    return HitlDecision(False, reason="no_actionable_command_detected")
```

## 5. Query

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

_pending_approval = None


def query(question):
    global _pending_approval
    template = """ì§ˆë¬¸ì— ë‹µë³€í•˜ì„¸ìš”.

ì§ˆë¬¸: {question}

ë‹µë³€:"""
    prompt = ChatPromptTemplate.from_template(template)
    llm_chain = prompt | llm | StrOutputParser()
    response_text = llm_chain.invoke({"question": question})

    decision = decide_hitl(response_text)
    if decision.requires_approval:
        _pending_approval = {
            "proposed_response": response_text,
            "reason": decision.reason,
        }
        print("=" * 50)
        print("âš ï¸  Human-in-the-Loop ìŠ¹ì¸ í•„ìš”")
        print("=" * 50)
        print(f"\nì œì•ˆëœ ì‘ë‹µ:\n{response_text}\n")
        print("=" * 50)
        print("ì´ ì‘ë‹µì—ëŠ” ë¡œë´‡ì—ê²Œ ì‹¤ì œ ëª…ë ¹/ì‹¤í–‰ì„ ìœ ë„í•˜ëŠ” ë‚´ìš©ì´ í¬í•¨ë˜ì–´ ìˆìŠµë‹ˆë‹¤.")
        print("ìŠ¹ì¸í•˜ë ¤ë©´: approve()")
        print("ê±°ì ˆí•˜ë ¤ë©´: reject()")
        print("=" * 50)
        return {
            "status": "approval_required",
            "reason": decision.reason,
            "proposed_response": response_text,
        }

    return {
        "status": "ok",
        "response": response_text,
    }


def approve():
    """ìŠ¹ì¸ ëŒ€ê¸° ì¤‘ì¸ ì‘ë‹µì„ ìŠ¹ì¸í•©ë‹ˆë‹¤."""
    global _pending_approval
    if _pending_approval is None:
        print("ìŠ¹ì¸ ëŒ€ê¸° ì¤‘ì¸ ì‘ë‹µì´ ì—†ìŠµë‹ˆë‹¤.")
        return None
    
    approved_response = _pending_approval["proposed_response"]
    _pending_approval = None
    print("âœ… ìŠ¹ì¸ë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ìŒ ë‹¨ê³„ë¡œ ì§„í–‰í•©ë‹ˆë‹¤.")
    return {
        "status": "approved",
        "response": approved_response,
    }


def reject():
    """ìŠ¹ì¸ ëŒ€ê¸° ì¤‘ì¸ ì‘ë‹µì„ ê±°ì ˆí•©ë‹ˆë‹¤."""
    global _pending_approval
    if _pending_approval is None:
        print("ìŠ¹ì¸ ëŒ€ê¸° ì¤‘ì¸ ì‘ë‹µì´ ì—†ìŠµë‹ˆë‹¤.")
        return None
    
    _pending_approval = None
    print("âŒ ê±°ì ˆë˜ì—ˆìŠµë‹ˆë‹¤.")
    return {
        "status": "rejected",
    }
```
